```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import random
import numpy as np
%matplotlib inline
```

```python
class Particle:
  def __init__(self, x, y):
    self.x = x
    self.y = y
    self.fitness = eggHolder(x,y)
    self.velocity=np.array([0,0])
    self.pibest=np.array([x,y])
```

```python
def eggHolder(x,y):
    return (-(y+47)*np.sin(np.sqrt(np.abs((x/2)+(y+47))))- (x*np.sin(np.sqrt(np.abs(x-y-47)))))
```

```python
def holderTable(x,y):
    sxcy = np.sin(x)*np.cos(y)
    expArity = (np.sqrt(x**2 + y**2))/np.pi
    expArity = np.exp(np.abs(1-expArity))
    return -(np.abs(sxcy*expArity))
```

```python
# Constraints Given
EggHolderXMin = -512
EggHolderXMax = 512
EggHolderYMin = -512
EggHolderYMax = 512
HolderTableXMin = -10
HolderTableXMax = 10 
HolderTableYMin = -10
HolderTableYMax = 10 
```

```python
# Defining Required Arrays
Population=[]
Averages=[]
gBest = []
bestFitness=[]
```

```python
# Variable Params
VMax = 2

# c1+c2=4
c1 = 1
c2 = 3

NumberOfIterations = 20
PopulationSize = 10
```

```python
# Defining Required Funcs
def getVelocity(particle, VMax):
    middleTerm = c1*random.random()*(pbest-particle.pibest)
    endTerm = c2*random.random()*(gbest-particle.pibest)
    calculatedVelocity=particle.velocity+middleTerm+endTerm
    particle.velocity=particle.velocity+middleTerm+endTerm
    return particle.velocity


def getNewPosition(particle,Xmax, Xmin, Ymax, Ymin):
    x,y = getVelocity(particle, VMax)
    if((Xmin <= (x+particle.x) <= Xmax) and (Ymin <= (y+particle.y) <= Ymax)):
        particle.x+=x
        particle.y+=y
    else:
        tempX = particle.x+x
        tempY = particle.y+y
        if(tempX < Xmin):
            particle.x=Xmin
        else:
            particle.x = Xmax
        if(tempY < Ymin):
            particle.y=Ymin
        else:
            particle.y = Ymax
        particle.velocity=np.array([0,0])
        
    return x,y
```

```python
for particleNumber in range(PopulationSize):
    Population.append(Particle(random.uniform(EggHolderXMin,EggHolderXMax),random.uniform(EggHolderYMin,EggHolderYMax)))
fitneessValues=[]
for particle in Population:
    fitneessValues.append(particle.fitness)
fbest, pbest = min(fitneessValues),np.array([Population[fitneessValues.index(min(fitneessValues))].x,Population[fitneessValues.index(min(fitneessValues))].y])
gbest=pbest
for i in range(NumberOfIterations):
    for particle in Population:
        getNewPosition(particle, EggHolderXMax, EggHolderXMin, EggHolderYMax, EggHolderYMin)
    for particle in Population:
        particle.fitness= eggHolder(particle.x,particle.y) 
    fbest1=fbest
    pbest1=pbest
    fitneessValues=[]
    for particle in Population:
        if(particle.fitness < eggHolder(particle.pibest[0],particle.pibest[1])):
            particle.pibest = np.array([particle.x,particle.y])
        fitneessValues.append(particle.fitness)
    Averages.append((sum(fitneessValues)/len(fitneessValues)))
    fbest, pbest = min(fitneessValues),np.array([Population[fitneessValues.index(min(fitneessValues))].x,Population[fitneessValues.index(min(fitneessValues))].y])
    if fbest<=fbest1:
        gBest.append(gbest)
        bestFitness.append(eggHolder(gbest[0],gbest[1]))
        gbest=pbest   
```
